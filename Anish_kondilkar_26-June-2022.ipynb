## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

Created a new DataFrame 'df' and 'copy' is used to copy data in new dataframe.

```python
df=ECom_Data.copy()
```

'l3' is used to define a new column which is being created by using if,elif & else conditons to fetch values from 'discounted_price' column. where if 'j'in discounted_price is greater then given order(in percentage), it will multiply it by 'j' and append.

```python
l3=[]
for j in df['discounted_price']:
    if (j>600):
        l3.append(0.25*j)
    elif (j>350) & (j<=600):
        l3.append(0.15*j)
    elif (j>100) & (j<=350):
        l3.append(0.1*j)
    else:
        l3.append(0.05*j)
df['E-Commerce Revenue']=l3
```

```python
df.groupby('Brand').sum()
```

## Q6.Calculate the BrandRevenue for each brand and list the top 14 brands having maximum revenue in descending order (6 marks)
#### Hint: Brand revenue is different from the E-Commerce company revenue.


### Ans 6

'l' is used to define a new column which is being created by using if,elif conditons to fetch values from 'discounted_price' column. where if 'i'in discounted_price is greater then given order(in percentage), it will multiply it by 'i' and append. we have asummed 'Brand Revenue'as, if company charges 25% on final value(discounted value) then the brand revenue will get 75% of the final value.

```python
l=[]
for i in df['discounted_price']:
    if (i>600):
        l.append(0.75*i)
    elif (i>350) & (i<=600):
        l.append(0.85*i)
    elif (i>100) & (i<=350):
        l.append(0.9*i)
    elif (i<=100):
        l.append(0.95*i)
df['brand_revenue']=l
```

```python
df.groupby('Brand')[['brand_revenue']].sum()
```

```python
df.groupby('Brand').sum().sort_values(by='brand_revenue',ascending=False)['brand_revenue'].head(14)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of ECom Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

'Order date' data type is in object datatype format which is  converted to datatime datatype so it will help in extracting the information of month and year.

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'],dayfirst=True)
```

```python
df.info()
```

In cell 45 & 46 we are extracting data of year & month from 'Order Date' column

```python
df['year']=df['Order_Date'].dt.year
```

```python
df['month']=df['Order_Date'].dt.month
```

Here we can see that separate month and year column has been created.

```python
df.head()
```

By using 'strftime'is used to set string format of time in 'Month & Year' column.

```python
df['Month & Year']=df['Order_Date'].dt.strftime('%m/%Y')
df.head()
```

Here new Dataframe'df5' is made to add E-commerce Revenue column.

```python
df5=df.groupby(['Region','year','month','Month & Year'])[['E-Commerce Revenue']].sum().copy()
df5
```

```python
df5.reset_index(inplace=True)
df5
```

Here data is extracted from 'df5' dataframe where hue is used for colours and marker is used for shape of the points. 

```python
plt.figure(figsize=(22,11))
sns.lineplot(data=df5,x='Month & Year',y='E-Commerce Revenue',hue='Region',marker='o')
plt.xlabel('Month & Year',fontsize=15)
plt.ylabel('E-commerce revenue',fontsize=15)
plt.show()
```

### Ans 9 b)

'df6' shows the Best performing E-Commerce Revenue on basis of region.

```python
df6= df5.groupby(['Region'])[['E-Commerce Revenue']].max().reset_index()
df6
```

'df7' shows the Best performing E-Commerce Revenue on basis of region.

```python
df7= df5.groupby(['Region'])[['E-Commerce Revenue']].min().reset_index()
df7
```

'df8' new dataframe is created where groupby is done to add month & year column.

```python
df8=df5.groupby(['Region','Month & Year']).sum().copy().reset_index()
df8
```

'df8' is merged with maximum E-Commerce revenue dataframe 'df6' to get Best performing month for each region.

```python
pd.merge(df8,df6,on=['Region','E-Commerce Revenue'])
```

'df8' is merged with maximum E-Commerce revenue dataframe 'df7' to get worst performing month for each region.

```python
pd.merge(df8,df7,on=['Region','E-Commerce Revenue'])
```
